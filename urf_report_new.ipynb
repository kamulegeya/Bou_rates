```python
##use a function to get Voucher number from WHT
import re
def get_inv_num(string):
    try:
        pattern = re.findall(r'RM\d+',string)
        return ",".join(pattern)
    except:
        return 'err'
```

```python
df_wht['original_inv_num']=df_wht['Description'].apply(get_inv_num)
```
