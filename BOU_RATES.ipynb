## PYTHON SCRIP TO WEB SCRAP  BOU EXCHANGE RATES


Recently bank of Uganda only allows you to down load data for a single day by entering data using the parameters of year, month and day.

```python
import pandas as pd
import ssl
dd=pd.DataFrame()
ssl._create_default_https_context = ssl._create_unverified_context
months=['Jul','Aug','Sep']
for m in months:
    for i in range(1,32):
        strday=str(i)
        ## days are written as two digits...so we append a value of 0 if less than 10
        if len(strday)!=2:
            Link = 'https://www.bou.or.ug/bou/bouwebsite/bouwebsitecontent/ExchangeRates/scripts.MajorExchangeRates/index.jsp?year=2021&month='+m+'&day='+str('0')+str(i)
        else:
            Link = 'https://www.bou.or.ug/bou/bouwebsite/bouwebsitecontent/ExchangeRates/scripts.MajorExchangeRates/index.jsp?year=2021&month='+m+'&day='+str(i)
        try:
            table_MN = pd.read_html(Link)
            dd=dd.append(table_MN[0])
                      
                                   
        except Exception as e:
            print(e)
dd
```

```python
# Get names of indexes for which column DATE COLUMN HAS VALUE DATE
indexNames = dd[dd['DATE'] == 'DATE' ].index
# Delete these row indexes from dataFrame
dd.drop(indexNames , inplace=True)
```

```python
dd
```

```python
## convert date column to datetime
dd['DATE']=pd.to_datetime(dd['DATE'])
```

```python
##dd.to_excel("exchange_rates.xlsx")
```

```python
## ADD COLLUMN OF CUSTOME FORMAT MONTH
dd['Month'] = dd['DATE'].dt.strftime('%Y-%m')
            
```

```python
dd['SELLING RATE(UGX)'] = pd.to_numeric(dd['SELLING RATE(UGX)'],errors = 'coerce')
```

```python
dd['BUYING RATE (UGX)'] = pd.to_numeric(dd['BUYING RATE (UGX)'],errors = 'coerce')
```

```python
avg_agg = dd.groupby(['Month','CURRENCY']).agg({'SELLING RATE(UGX)': ['mean', 'min', 'max']},{'BUYING RATE(UGX)': ['mean', 'min', 'max']})
```
